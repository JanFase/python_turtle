```python
import turtle
import random
from sympy import Polygon

ond1h = 0
ond1v = 0

ond2h = 40
ond2v = 0

ond3h = 120
ond3v = 0

ond4h = 160
ond4v = 0

min_area = 160 * 140 * 60
max_area = 160 * 140 * 90

win_width, win_height, bg_color = 2000, 2000, 'white'

turtle.setup()
turtle.screensize(win_width, win_height, bg_color)

window = turtle.Screen()
window.bgcolor("white")

tegel = turtle.Turtle()
tegel.shape("turtle")
tegel.color("black")

factor = 1

base_h = 0 #-300
base_v = 0 #-300

adj_v = 0

#tegel.penup()
#tegel.goto(base_h, base_v)
#tegel.pendown()
```

```python
tegelNr = 1

while tegelNr < 11:
    
    print('')
    print('tegelNr =', tegelNr)

    nogNiet = True
    aantalRandom = 0
    while (nogNiet):
        #print('aantalRandom =', aantalRandom)
        bvn1h = 0
        bvn1v = random.randrange(  70, 130,  5)

        bvn2h = random.randrange(  20, 100,  5)
        bvn2v = random.randrange(  70, 130,  5)

        bvn3h = random.randrange(  bvn2h + 20, 140,  5)
        bvn3v = random.randrange(  70, 130,  5)

        bvn4h = 160
        bvn4v = random.randrange(  70, 130,  5)

        if(abs(bvn2v - bvn1v) <= 20) or (abs(bvn3v - bvn2v) <= 20) or (abs(bvn4v - bvn3v) <= 20):
            nogNiet = True
            #print(abs(bvn2v - bvn1v))
            #print(abs(bvn3v - bvn2v))
            #print(abs(bvn4v - bvn3v))
        else:
            nogNiet = False
 
            prtv1 = bvn1v
            bvn1v += adj_v
            
            prtv2 = bvn2v
            bvn2v += adj_v

            prtv3 = bvn3v
            bvn3v += adj_v

            prtv4 = bvn4v
            bvn4v += adj_v
            
        #print(nogNiet)
        #print('')
        aantalRandom += 1

    areaTegel = Polygon((ond1h, ond1v), (ond2h, ond2v), (ond3h, ond3v), (ond4h, ond4v),
                (bvn4h, bvn4v), (bvn3h, bvn3v), (bvn2h, bvn2v), (bvn1h, bvn1v)
               ).area
    areaPerc = areaTegel / (160 * 140) * 100
    print('areaTegel = ' , int(areaTegel), 'area_perc = ', areaPerc)
    
    print('ga printen!')

    print('bvn1 = ', bvn1h, bvn1v)
    print('bvn2 = ', bvn2h, bvn2v)
    print('bvn3 = ', bvn3h, bvn3v)
    print('bvn4 = ', bvn4h, bvn4v)
    print('ond1 = ', ond1h, ond1v)
    print('ond2 = ', ond2h, ond2v)
    print('ond3 = ', ond3h, ond3v)
    print('ond4 = ', ond4h, ond4v)


    tegel.goto(  (ond1h * factor),   (ond1v * factor))
    tegel.pendown()
    tegel.goto(base_h + (bvn1h * factor), base_v + (bvn1v * factor))
    tegel.goto(base_h + (bvn2h * factor), base_v + (bvn2v * factor))
    tegel.goto(base_h + (bvn3h * factor), base_v + (bvn3v * factor))
    tegel.goto(base_h + (bvn4h * factor), base_v + (bvn4v * factor))

    tegel.goto(  (ond4h * factor),   (ond4v * factor))
    tegel.goto(  (ond3h * factor),   (ond3v * factor))
    tegel.goto(  (ond2h * factor),   (ond2v * factor))
    tegel.goto(  (ond1h * factor),   (ond1v * factor))
    
    tegel.penup()

    #add some text
    ond_avg = (ond1v + ond2v + ond3v + ond4v) / 4
    bvn_avg = (bvn1v + bvn2v + bvn3v + bvn4v) / 4
    pos_text = (bvn_avg + ond_avg) / 2
    print ('pos_text =', pos_text)
    
    tegel.penup()
    tegel.goto(200, pos_text)
    tegel.pendown()
    txt = 'Tegel ' +  str(tegelNr) + '\n' + \
              str(bvn1h) + ' ' + str(prtv1)  + ' | ' + \
              str(bvn2h) + ' ' + str(prtv2)  + ' | ' + \
              str(bvn3h) + ' ' + str(prtv3)  + ' | ' + \
              str(bvn4h) + ' ' + str(prtv4) +  '\n' + \
            'opp = ' + str(int(areaTegel)) + ' opp % = ' + str(int(areaPerc))
    tegel.write(txt, move=False, align='left', font=('Verdana', 7, 'normal')) 
    tegel.penup()

    ond1h = bvn1h 
    ond1v = bvn1v + 16
    
    ond2h = bvn2h 
    ond2v = bvn2v + 16
    
    ond3h = bvn3h 
    ond3v = bvn3v + 16
    
    ond4h = bvn4h 
    ond4v = bvn4v + 16

    ond_set = ([bvn1v, bvn2v, bvn3v, bvn4v])
    print('min onder  = ', min(ond_set))
    adj_v =  16 + min(ond_set)


    tegelNr += 1

tegel.penup()
tegel.goto(200, 0)
```
